### Q1 :- Print the given strings as per stated format.

```

"By" "CampusX"
```

``` 
Data-Science-Mentorship-Program-started-By-CampusX
```

### Q2:- Write a program that will convert celsius value to fahrenheit.

````python
celsius = float(input('Temperature in Celsius: '))
fahrenheit = (celsius * 9/5) + 32
print(f"Temperature in fahrenheit is: {fahrenheit}.")
````

### Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.

````python
f_num = int(input("Enter your first number: "))
s_num = int(input("Enter your second number: "))

temp = f_num
f_num = s_num
s_num = temp

print(f"f_num is {f_num}")
print(f"s_num is {s_num}")
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Taking input for the first coordinate (x1, y1)
x1 = float(input("Enter x1: "))
y1 = float(input("Enter y1: "))

# Taking input for the second coordinate (x2, y2)
x2 = float(input("Enter x2: "))
y2 = float(input("Enter y2: "))

# Calculating the Euclidean distance without math module
distance = ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5

# Displaying the result
print(f"The Euclidean distance between the two points is: {distance:.2f}")
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
# Taking input from the user
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the rate of interest (in %): "))
time = float(input("Enter the time period (in years): "))

# Calculating simple interest
simple_interest = (principal * rate * time) / 100

# Displaying the result
print(f"The Simple Interest is: {simple_interest:.2f}")
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2


````python
# Program to find the number of dogs and chickens
# based on total heads and legs

# Input from user
heads = int(input("Enter total number of heads: "))
legs = int(input("Enter total number of legs: "))



dogs = (legs - 2 * heads) // 2
chickens = heads - dogs

# Check for validity
if dogs < 0 or chickens < 0 or legs % 2 != 0:
    print("No valid solution exists.")
else:
    print(f"Dogs -> {dogs}")
    print(f"Chickens -> {chickens}")
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
n = int(input("Enter a number: "))

result = (n*(n+1)*(2*n + 1)) / 6
print(result)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
first_term = int(input("Enter the first term: "))
second_term = int(input("Enter the second term: "))
n = int(input("Enter the value of n"))

d = second_term - first_term

an = first_term + (n - 1) * d
print(an)

````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
n1 = int(input("Enter num1: "))
d1 = int(input("Enter den1: "))
n2 = int(input("Enter num2: "))
d2 = int(input("Enter den2: "))

rn = n1*d2 + n2*d1
rd = d1 * d2

print('{}/{}'.format(rn, rd))



````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
import math

# Input values
H = float(input("Enter height of tank (cm): "))
L = float(input("Enter length of tank (cm): "))
B = float(input("Enter breadth of tank (cm): "))

h = float(input("Enter height of glass (cm): "))
r = float(input("Enter radius of glass (cm): "))

# Volume calculations
tank_volume = H * L * B
glass_volume = math.pi * (r ** 2) * h

# Number of glasses
num_glasses = tank_volume / glass_volume

print(f"\nNumber of glasses of milk that can be obtained: {int(num_glasses)}")
````
