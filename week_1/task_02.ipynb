### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
ctc = int(input('Enter your anual CTC: '))

if ctc < 500000:
    salary = ctc * 0.82
elif ctc < 1000000:
    salary = ctc * 0.72
elif ctc < 2000000:
    salary = ctc * 0.62
else:
    salary = ctc * 0.52

print(f"your in hand monthly salary will be: {round(salary/12, 2)}")
````

### `Problem 2`: Write a program that take a user input of three angles and will find out whether it can form a triangle or not.

````python
first = int(input("Enter first angle: "))
second = int(input("Enter second angle: "))
third = int(input("Enter third angle: "))

if (first + second + third) == 180 and first > 0 and second > 0 and third > 0:
    print("forms a triangle.")
else:
    print("does not form a triangle.")
````

### `Problem 3`: Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit.

````python
cost_price = int(input("Enter your cost price: "))
sell_price = int(input("Enter your selling price: "))

if cost_price > sell_price:
    print('Profit')
elif sell_price > cost_price:
    print('Loss')
else:
    print('Break even.')
````

### `Problem 4`: Write a menu-driven program - 
1. cm to ft 
2. km to miles 
3. USD to INR 
4. exit

````python
menu = int(input("""
Select an option:
1. cm to ft
2. km to miles
3. USD to PKR
4. exit.
"""))

if menu == 1:
    cm = float(input('Cm Value: '))
    ft = cm * 0.32
    print(f"ft value is {ft}.")
elif menu == 2:
    km = float(input('km Value: '))
    mile = km * 0.62
    print(f"miles value is {mile}.")
elif menu == 3:
    usd = float(input('USD Value: '))
    pkr = usd * 300
    print(f"PKR value is {pkr}.")
else:
    exit()
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output: 

```bash
120
```

````python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num+1):
    fact = fact * i

print(f"The factorial is {fact}")
````

````python
num = int(input("Enter the number: "))
rev = 0

while num > 0:
    last = num % 10
    rev = rev * 10 + last
    num = num // 10

print(rev)
````
