### Q1: Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)] 
Output : [(5, 6, 7, 8), (6, 10), (7, 13)] 
```

````python
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
unique = []

for i in test_list:
    unique.append(i[0])
unique = set(unique)

result = []
for i in unique:
    result.append([i])
    for j in test_list:
        if j[0] == i:
            result[-1].append(j[1])
            
print(result)
````

### Q2 Multiply Adjacent elements (both sides) and take the sum of the right and lest side multiplication results. 


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication : 

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
t = (1, 5, 7, 8, 10)

L = []
L.append(t[0] * t[1])


for i in range(1, len(t) -1):
    L.append(t[i] * t[i - 1] + t[i] * t[i + 1])

L.append(t[-1] * t[-2])
print(L)
````

### Q4: Count no of tuples, lists, and sets from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
L = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

output = [0,0,0]
for i in L:
    if type(i) == list:
        output[0] += 1
    elif type(i) == set:
        output[1] += 1
    elif type(i) == tuple:
        output[2] += 1
    else:
        pass

print(f"List-{output[0]}\nSet-{output[1]}\nTuples-{output[2]}")
````

### Q5: Shortlist Students for a Job Role
Ask the user to input students' records and store them in tuples for each record. Then ask the user to input three things he wants in the candidate- Primary Skill, Higher Education, Year of Graduation.

Show every students record in form of tuples if matches all required criteria.

It is assumed that there will be only one primry skill.

If no such candidate found, print `No such candidate`

`Input:`
```
Enter No of records- 2
Enter Details of student-1
Enter Student name- Manohar
Enter Higher Education- B.Tech
Enter Primary Skill- Python
Enter Year of Graduation- 2022
Enter Details of student-2
Enter Student name- Ponian
Enter Higher Education- B.Sc.
Enter Primary Skill- C++
Enter Year of Graduation- 2020

Enter Job Role Requirement
Enter Skill- Python
Enter Higher Education- B.Tech
Enter Year of Graduation- 2022
```

`Output`
```
('Manohar', 'B.tech', 'Python', '2022')
```


````python
# Number of student records
n = int(input("Enter No of records- "))

students = []

# Collect each student's details
for i in range(1, n + 1):
    print(f"Enter Details of student-{i}")
    name = input("Enter Student name- ")
    education = input("Enter Higher Education- ")
    skill = input("Enter Primary Skill- ")
    grad_year = input("Enter Year of Graduation- ")
    
    # Store each record as a tuple
    students.append((name, education, skill, grad_year))

# Job role requirement
print("\nEnter Job Role Requirement")
req_skill = input("Enter Skill- ")
req_education = input("Enter Higher Education- ")
req_year = input("Enter Year of Graduation- ")

# Filtering candidates
matched = []

for s in students:
    if s[1] == req_education and s[2] == req_skill and s[3] == req_year:
        matched.append(s)

# Output result
if matched:
    for m in matched:
        print(m)
else:
    print("No such candidate")
````

### Q1: Write a program to find a set of common elements in three lists using sets.
```
Input : ar1 = [1, 5, 10, 20, 40, 80]
        ar2 = [6, 7, 20, 80, 100]
        ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

Output : [80, 20]
```

````python
ar1 = set([1, 5, 10, 20, 40, 80])
ar2 = set([6, 7, 20, 80, 100])
ar3 = set([3, 4, 15, 20, 30, 70, 80, 120])

result = (ar1 & ar2) & ar3
print(list(result))
````

### Q4: find union of n arrays.

**Example 1:**

Input:
```bash
[[1, 2, 2, 4, 3, 6],
 [5, 1, 3, 4],
 [9, 5, 7, 1],
 [2, 4, 1, 3]]
```

Output:


[1, 2, 3, 4, 5, 6, 7, 9]
